# Creating Movie Recomendation

```python
df.head()
```

```python
movie_mat = df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
movie_mat
```

```python
star_wars_rating = movie_mat['Star Wars (1977)']
star_wars_rating.head()
```

```python
similar_to_starWars = movie_mat.corrwith(star_wars_rating)
```

```python
corr_to_starwars = pd.DataFrame(similar_to_starWars,columns=['Correlation'])
```

```python
corr_to_starwars.dropna(inplace=True)
```

```python
corr_to_starwars.head()
```

```python
corr_to_starwars.sort_values('Correlation' , ascending=False).head(10)
```

```python
corr_to_starwars = corr_to_starwars.join(ratings['no of ratings'])
```

```python
corr_to_starwars[corr_to_starwars['no of ratings']>50].sort_values(by='Correlation',ascending=False)
```

## Prediction Function

```python
def predict_movies(movie_name) :
    try :
        movie_user_rating = movie_mat[movie_name]
        similar_to_movie = movie_mat.corrwith(movie_user_rating)
        
        corr_to_movie = pd.DataFrame(similar_to_starWars,columns=['Correlation'])
        corr_to_movie.dropna(inplace=True)
        
        corr_to_movie = corr_to_movie.join(ratings['no of ratings'])
        recommendations =  corr_to_movie[corr_to_movie['no of ratings']>50].sort_values(by='Correlation',ascending=False)
        return recommendations
    except Exception as e :
        print(e)
        
```

```python
recommendations = predict_movies('Titanic (1997)')
try :
    for i in range (1,11) :
        print(recommendations.index[i])
except Exception :
    print('You have not entered correct movie name' )
```
